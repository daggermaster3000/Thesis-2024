```python
import matplotlib.pyplot as plt
import pandas as pd

pkd2l1_total = 733
cilia_total = 153

data = {'Colocalized': [89/pkd2l1_total, 89/cilia_total], 'Non colocalized': [1-89/pkd2l1_total, 1-89/cilia_total]}
df = pd.DataFrame(data, index=['Pkd2l1 puncta', 'Actub puncta'])

# Use ggplot style
plt.style.use('ggplot')

ax = df.plot(kind='barh', stacked=True, color=['limegreen', 'grey'], figsize=(10,3))
plt.xticks(rotation=0)  # Rotate x-axis labels to be horizontal
plt.xlabel('Percentage')
plt.ylabel('Categories')
plt.title('Detected Pkd2l1 Puncta Colocalization')

plt.show()
```
